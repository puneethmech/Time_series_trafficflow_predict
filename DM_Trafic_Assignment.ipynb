```python
## the is for plotting df_tra_Y_tr data for first row

#
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

data_df_00 = df_tra_Y_te.iloc[1:2]
data_df_00_t = data_df_00.transpose()

##### get row number of the dataframe and insert it as first column
 
data_df_00_t.insert(loc=0, column='row_num', value=np.arange(len(data_df_00_t)))
# adding column name to the respective columns
data_df_00_t.columns =['Time_indx', 'Traffic_flow']


data_df_00_t.set_index('Time_indx',inplace=True)

sensornumber = "sensortest0"

from pylab import rcParams
rcParams['figure.figsize'] = 15, 7
graph = data_df_00_t.plot()

fig = graph.get_figure()
#fig.savefig("output.png")
fig.savefig(sensornumber+".jpeg")

```

```python
## adding adfuller to chek if the 

from statsmodels.tsa.stattools import adfuller

test_result=adfuller(data_df_00_t['Traffic_flow'])

def adfuller_test(Traffic_flow):
    result = adfuller(Traffic_flow)
    labels = ['ADF Test Statistic','p-value','#Lags Used','Number of Observations']
    for value,label in zip(result,labels):
        print(label+' : '+str(value) )

    if result[1] <= 0.05:
        print("strong evidence against the null hypothesis(Ho), reject the null hypothesis. Data is stationary")
    else:
        print("weak evidence against null hypothesis,indicating it is non-stationary ")

adfuller_test(data_df_00_t['Traffic_flow'])
```

```python
# For non-seasonal data
#p=1, d=1, q=0 or 1
from statsmodels.tsa.arima_model import ARIMA

model=ARIMA(data_df_00_t['Traffic_flow'],order=(1,1,1))
model_fit=model.fit()
model_fit.summary()
results=model.fit()
```

```python
import statsmodels.api as sm
model=sm.tsa.statespace.SARIMAX(data_df_00_t['Traffic_flow'],order=(1, 1, 1),seasonal_order=(1,1,1,96))
results=model.fit()
results.summary()
```

```python
data_df_00_t['forecast']=results.predict(start=400,end=839,dynamic=True)
data_df_00_t[['Traffic_flow','forecast']].plot(figsize=(12,8))
```
